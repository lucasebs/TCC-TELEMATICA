```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = pd.DataFrame()
for i in range(0,7):
    if (i == 3) or (i == 4): continue
    data = pd.read_csv('src/output/distributed/log/pti_tpi'+str(i)+'.txt', sep=";", header=None )
    data.columns = ["image", "time"]
    
    data_sorted = data.sort_values(["image"])
    d[str(i)] = data_sorted["time"]
    d = d.set_index(data_sorted["image"])
```

```python
d.head(25).to_csv('distributed_tpi.csv')
d.head(25)
```

```python
aux = d.transpose().describe().transpose()
aux["count"] = aux['count'].astype(int)
aux["min"] = aux['min'].astype(int)
aux["25%"] = aux['25%'].astype(int)
aux["50%"] = aux['50%'].astype(int)
aux["75%"] = aux['75%'].astype(int)
aux["max"] = aux['max'].astype(int)
aux_2 = d.transpose().aggregate([np.median]).transpose().astype(int)
d_final = pd.concat([aux,aux_2], axis=1)
d_final_describe = d_final['mean'].describe().to_frame().transpose()
d_final.columns = ["qtd","media","std","min",'25',"50","75","max",'mediana']
d_final = d_final.round(2)
d_final.head(25).to_csv('distributed.csv')
d_final.head(25)
```

```python
c = pd.DataFrame()
for i in range(0,7):
    if (i == 3) or (i == 4): continue
    data_c = pd.read_csv('src/output/centralized/log/pti_tpi'+str(i)+'.txt', sep=";", header=None)
    data_c.columns = ["image", "time"]
    
    data_c["image"] = data_c["image"].str.split("_",n=1,expand=True)[0].astype(int)
    data_c_sorted = data_c.sort_values(["image"])    
    c[str(i)] = data_c_sorted["time"]
    c = c.set_index(data_c_sorted["image"])
```

```python
c.head(25).to_csv('centralized_tpi.csv')
c.head(25)
```

```python
aux = c.transpose().describe().transpose()
aux["count"] = aux['count'].astype(int)
aux["min"] = aux['min'].astype(int)
aux["25%"] = aux['25%'].astype(int)
aux["50%"] = aux['50%'].astype(int)
aux["75%"] = aux['75%'].astype(int)
aux["max"] = aux['max'].astype(int)
aux_2 = c.transpose().aggregate([np.median]).transpose().astype(int)
c_final = pd.concat([aux,aux_2], axis=1)
c_final_describe = c_final['mean'].describe().to_frame().transpose()
c_final.columns = ["qtd","media","std","min",'25',"50","75","max",'mediana']
c_final = c_final.round(2)
c_final.head(25).to_csv('centralized.csv')
c_final.head(25)
```

```python
finals_describe = [d_final_describe, c_final_describe]
df_describe = pd.concat(finals_describe,ignore_index=True, axis=0)
df_describe.index = ['distributed', 'centralized']
df_describe = df_describe.round(2)
df_describe.columns = ["qtd",'media','std','min','25','50','75','max']
df_describe.to_csv('describe.csv')
df_describe
```
